```python
fake = pd.read_csv(r'C:\Users\kartik.nerwal\Downloads\archive/Fake.csv')
real = pd.read_csv(r'C:\Users\kartik.nerwal\Downloads\archive/True.csv')
```

```python
fake['Class'] = 0
real['Class'] = 1
```

```python
df = pd.DataFrame()
df = pd.concat([df,fake,real])
```

```python
df = df.dropna()
```

```python
X = df['title']
y = df['Class']
```

```python
X.shape
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
vocab_size = 5000
```

```python
messages = X.copy()
```

```python
messages.reset_index(inplace=True, drop=True)
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages[i])
    review = review.lower()
    review = review.split()

    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
onehot_repr = [one_hot(words, vocab_size) for words in corpus]
```

```python
onehot_repr
```

```python
sent_length = 20
embedding_docs = pad_sequences(onehot_repr, padding='pre', maxlen = sent_length)
```

```python
embedding_docs
```

```python
len(embedding_docs)
```

```python
embedding_docs[0]
```

```python
embedding_vector_features = 40
model = Sequential()
model.add(Embedding(vocab_size, embedding_vector_features, input_length=sent_length))
model.add(Dropout(0.3))
model.add(LSTM(100))
model.add(Dropout(0.3))
model.add(Dense(100, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
final_X = np.array(embedding_docs)
final_y = np.array(y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(final_X, final_y, test_size=0.3, random_state=0)
```

```python
history = model.fit(X_train, y_train, validation_data = (X_test, y_test), epochs=10, batch_size=64)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
y_pred
```

```python
y_pred = [1 if i>=0.5 else 0 for i in y_pred ]
```

```python
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```
